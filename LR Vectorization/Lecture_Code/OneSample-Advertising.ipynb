```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def predict(x, theta):
    return x.T.dot(theta)

def compute_loss(y_hat, y):
    return (y_hat-y)**2

def gradient_compute(x, y_hat, y):
    return 2*x*(y_hat-y)
def update_paramter(theta, gradient, lr):
    return theta-lr*gradient
```

# 1. Load data

```python
df = pd.read_csv('../data/advertising.csv', header=None)
features=df.iloc[1:,:3].astype(float).to_numpy()
N=features.shape[0]
target=df.iloc[1:,3].astype(float).to_numpy()
```

```python
X=np.hstack((np.ones((N,1)), features))
X
```

# 2. Training

```python
thetas = np.random.randn(4)

lr = 0.00001
epoch_max = 50
losses=[]

for _ in range(epoch_max):
    for i in range(N):
        x=X[i]
        y=target[i]
        y_hat=predict(x, thetas)
        gradient=gradient_compute(x, y_hat, y)
        thetas=update_paramter(thetas, gradient, lr)
        losses.append(compute_loss(y_hat, y))

plt.plot(losses[:500])
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss over Iterations')
plt.show()
```
