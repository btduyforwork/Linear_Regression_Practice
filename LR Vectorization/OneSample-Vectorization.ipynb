```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 0. Build functions

```python
def predict(x, theta):
    return x.T.dot(theta)

def compute_loss(y_hat, y):
    return (y_hat-y)**2

def gradient_compute(x, y_hat, y):
    return 2*x*(y_hat-y)
def update_paramter(theta, gradient, lr):
    return theta-lr*gradient
```

```python
theta=np.array([0.0, -0.34, 0.3, 0.2])
x=np.array([1, 44, 45, 49])
y=10
y_hat=predict(x, theta)
print(compute_loss(y_hat, y))
gradient=gradient_compute(x, y_hat, y)
theta=update_paramter(theta, gradient, 0.0001)
```

# 1. Load data

```python
df = pd.read_csv('../data/data.csv', header=None)
areas=np.array(df.iloc[:, 0])
target=np.array(df.iloc[:,1])
print(areas.shape)
print(target.shape)
```

# 2. Training

```python
N = 4
areas=areas.reshape(1, N)
X = np.vstack([np.ones((1, N)), areas])
theta = np.array([0.049, -0.34])

lr = 0.01
epoch_max = 10
losses=[]

for _ in range(epoch_max):
    for i in range(N):
        x=X[:,i]
        y=target[i]
        y_hat=predict(x, theta)
        gradient=gradient_compute(x, y_hat, y)
        theta=update_paramter(theta, gradient, lr)
        losses.append(compute_loss(y_hat, y))

plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss over Iterations')
plt.show()
```

```python
x_data = range(2, 8)

predicted_b, predicted_w = theta
y_data = [x*predicted_w + predicted_b for x in x_data]
plt.plot(x_data, y_data)
#--------------------------------------
plt.scatter(areas, target)
plt.xlim(3,7)
plt.ylim(4,10)
plt.show()
```
